## Is there a need to segment customers?

Customer segmentation based on their potential value is still a relevant process concerning any business with a product to market. By determining where individuals lie on the value matrix we are able to label which users are our ‘champions’, ‘at risk of exit’, ‘new to the product’ and so on. Knowing the different cuts that users fall under help the business build different strategies for focused & personalized marketing, helps to lower marketing costs and better product development.

Another important use case for customer segmentation is smart marketing. Marketing customers for a product is usually based on their online activity. Did they search/purchase the product or something similar before? What is their geographical area? How old are our customers? These are still valid questions but not necessarily the the only criteria for running a targeted campaign. How do we ensure that an interested customer and a regular customer purchases the product and continues to do so? By segmenting customers based on their value we are able to create detailed targeted campaigns that have a clear objective. A regular customer who is loyal to your brand can be targeted with an ad to trial a new product while a new customer who has exhibited 'interest’ in the brand can be targeted with an ad to redeem a first time buyers offer. Customers should be targeted differently for effective campaign results. 

## How do we segment customers?

While there are different techniques to categorize customers the most commonly known method is called RFM Segmentation which stands for Recency, Frequency and Monetary Value. These three quantifiable metrics are key to determining a customer’s behaviour; frequency and monetary value gives an indication of what is the customer’s lifetime value while recency indicates retention and engagement of the customer. 

```python
Image('rfm_metrics.png')
```

RFM began during the days of direct marketing but to this day is still one of the simplest and easily implementable as well as effective tool in the retail and e-commerce industry. We will practise how to implement conduct segmentation of an online sales data with RFM and then with K-Means Clustering.

#### Step 3: Use RFM Model to Segment Customers
Filter for only the required columns: CustomerID, InvoiceDate, InvoiceNo, Quantity and UnitPrice. Then to calculate recency of the transaction, make the present date one date after the last transaction date.

```python
data = df.loc[:,['CustomerID','InvoiceDate','InvoiceNo','Quantity','UnitPrice']]
data['TotalSales'] = df['Quantity'] * df['UnitPrice']
data.head()
```

```python
present = data['InvoiceDate'].max() + timedelta(days=1)
present
```

For Recency, Calculate the number of days between present date and date of last purchase each customer.\
For Frequency, Calculate the number of orders for each customer.\
For Monetary, Calculate sum of purchase price for each customer.

```python
rfm = data.groupby('CustomerID').agg({'InvoiceDate': lambda date: (present - date.max()).days,
                                      'InvoiceNo': lambda num: len(num),
                                      'TotalSales': lambda price: price.sum()
                                     })
rfm.rename(columns={'InvoiceDate':'Recency','InvoiceNo':'Frequency','TotalSales':'Monetary'}, inplace = True)
rfm.head()
```

When graphing the density distributions of Recency, Frequency and Monetary it is to no surprise that all of the data is positively skewed. We can see clearly with the graphs below between what points the data is the highest percentage of distributions.

```python
plt.figure(figsize=(12,10))
plt.subplot(3,1,1); sns.distplot(rfm['Recency'])
plt.subplot(3,1,2); sns.distplot(rfm['Frequency'])
plt.subplot(3,1,3); sns.distplot(rfm['Monetary'])
plt.show()
```

##### Use Quantiles to determine the RFM Values:
Customers with the lowest recency, highest frequency and monetary amounts are considered top customers\
Using "qcut()" a quantile based discretization function we will bin the data into 4 quantiles

Customers who purchased recently, are frequent buyers and spend a lot are assigned score of 444 – Recency(R) – 4, Frequency(F) – 4, Monetary(M) – 4\
These are our best customers, we can now group customers together based on their score and identify their segments

```python
r_labels = range(4,0,-1)
f_labels = range(1,5)
m_labels = range (1,5)
rfm['r_quartile'] = pd.qcut(rfm['Recency'], 4, labels = r_labels)
rfm['f_quartile'] = pd.qcut(rfm['Frequency'], 4, labels = f_labels)
rfm['m_quartile'] = pd.qcut(rfm['Monetary'], 4, m_labels)
```

```python
rfm['RFM_Segment'] = rfm['r_quartile'].astype(str) + rfm['f_quartile'].astype(str) + rfm['m_quartile'].astype(str)
```

Assign an 'RFM Score' based on summing the quartile values, the higher the score, the better the customer 

```python
rfm['RFM_Score'] = rfm[['r_quartile','f_quartile','m_quartile']].sum(axis=1)
rfm.head()
```

Now that we've labeled our customers based their RFM Score we are able to take the score and arbitrarily group customers together by using quantiles. We can see based on the aggregation below that the 'Gold' segment has the lowest mean recency, the highest mean frequency and the highest mean monetary value but also the lowest customer count. This makes sense because our data was always positively skewed, meaning most customers had lower recency, frequency and monetary values.

```python
labels = ['Platinum','Bronze','Silver','Gold']
rfm['RFM_Level'] = pd.qcut(rfm['RFM_Score'], q=4, labels = labels)
stats = rfm.groupby('RFM_Level').agg({'Recency':'mean','Frequency':'mean','Monetary':['mean','count']})
stats
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
squarify.plot(sizes=stats.iloc[:,3],label=stats.index, color=["lightsteelblue","darkgoldenrod","silver", "gold"])
plt.title('Tree Map of Customers in Each Level')
plt.axis('off')
plt.show()
```

#### Step 4: Use K-Means Clustering to Segment Customers

K-Means clustering is an unsupervised machine learning algorithm. It essentially looks for a fixed number (k) of clusters in a dataset. k is the number of number of centroids (represents the center of each cluster) needed in the dataset and every data point is allocated to each cluster by reducing the linear sum of squares. Iterative calculations are performed to optimize the positions of the centriods.

Data Processing Steps:
1. Unskew the data - using log transformation
2. Standardize data
3. Choosing the Optimal Number of Clusters
4. Apply K-Means Algorithm

```python
Image('clustering2.png')
```

After Clustering - can see groups of like data

##### 1. Unskew data using Log Transformation

```python
rfm_log = rfm[['Recency', 'Frequency', 'Monetary']].apply(np.log, axis = 1).round(3)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(3,1,1); sns.distplot(rfm_log['Recency'])
plt.subplot(3,1,2); sns.distplot(rfm_log['Frequency'])
plt.subplot(3,1,3); sns.distplot(rfm_log['Monetary'])
plt.show()
```

##### 2. Scale the data with StandardScaler from the SKLearn Library

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix,classification_report
from sklearn import metrics 

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_log)
rfm_scaled
```

##### 3. Choosing the Optimal Number of Clusters

Before applying K-Means we need to define the optimal number of clusters (k) for the data set. The most common method without randomly choosing a number is known as the Elbow Method. The idea of the Elbow Method is to iterate k-means clustering on the data set for a range of values of k (ex. k from 1 to 10 times). For each value of k, the sum of squared errors (SSE) is calculated. Then plot a line chart of the SSE for each value of k. SSE decreases as we increase the number of clusters (k); SSE will become 0 when k is equal to the number of data points in the dataset. Our goal is to choose a small value of k that has a low SSE, the 'elbow' usually represents the point where we start to have diminishing returns by increasing k.

Elbow criterion method:
1. Plot the number of clusters against within-cluster sum-of-squared-errors (SSE) - sum of squared distances from every data point to their cluster center
2. Identify an "elbow" in the plot
3. Elbow - a point representing an "optimal" number of clusters

```python
from yellowbrick.cluster import KElbowVisualizer

K = range(1,11)

for k in K:
    kmeanModel = KMeans(n_clusters= k, random_state=101)
    kmeanModel.fit(rfm_scaled)
    
visualizer = KElbowVisualizer(kmeanModel, k=(1,11))
visualizer.fit(rfm_scaled)        
visualizer.show()       
plt.show()
```

The KElbowVisualizer implements the 'elbow' method to help select the optimal number of clusters. Here we can see the optimal number of clusters for our applied data set is 3

##### 4. Apply K-Means Algorithm

We can now apply the K-Means Algorithm to our dataset with n_clusters=3, then assign the cluster labels to our original dataset

```python
kmeanModel = KMeans(n_clusters= 3, random_state=101)
kmeanModel.fit(rfm_scaled)

rfm['K_Cluster'] = kmeanModel.labels_
rfm.head()
```

```python
plt.scatter(rfm_scaled[:, 0], rfm_scaled[:, 1],c=kmeanModel.labels_ , cmap='Set2')

centers = kmeanModel.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200);
```

```python
rfm.groupby('K_Cluster').agg({'Recency':'mean','Frequency':'mean','Monetary':['mean','count']})
```

We can now see three segments of customers identified by K-Means. Looking at the mean values for Recency, Frequency and Monetary when grouped by the K_Cluster we can make our own segment names with assumptions. 
For example, customers in K_Cluster 0 seem to be the 'Best' customers with the lowest recency (mean=12.96), highest frequency (mean=260.8) and highest monetary value (mean=$6,584.54).
